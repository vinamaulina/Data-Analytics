# Data Visualization using Python

# Matplotlib

Matplotlib is a cross-platform, data visualization and graphical plotting library (histograms, scatter plots, bar charts, etc) for Python and its numerical extension NumPy. Matplotlib is powerful tool for executing a variety of tasks. It is able to create different types of visualization reports like line plots, scatter plots, histograms, bar charts, pie charts, box plots, and many more different plots.

# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# Bar Graph

A bar graph is a graphical representation of information. It uses bars that extend to different heights to depict value. Bar graphs can be created with vertical bars, horizontal bars, grouped bars (multiple bars that compare values in a category), or stacked bars (bars containing multiple types of information).

```python
#using matplotlib
```

```python
plt.bar(years, tangerines) #plt.bar(x,y)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("crop yields in kanto")
```

```python
plt.bar(years, apples)
plt.bar(years, tangerines, bottom=apples)
```

similar to line chart, we can draw multiple histograms in a single chart. lets drew separate histograms for each species

```python
setosa_df = iris_df[iris_df.species == 'setosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha = 0.4)
plt.hist(versicolor_df.sepal_width, alpha = 0.4)
```

```python
flights_df = sns.load_dataset("flights")
flights_df.head()
```

```python
#specifying number of bins using numpy

import numpy as np
plt.hist(flights_df.passengers, bins=np.arange(110, 120, 2));
```

```python
jan_df = flights_df[flights_df.month == 'Jan']
feb_df = flights_df[flights_df.month == 'Feb']
mar_df = flights_df[flights_df.month == 'Mar']
```

```python
plt.hist(jan_df.passengers, alpha = 0.4)
plt.hist(feb_df.passengers, alpha = 0.4)
```

multiple histograms can be stacked on top of one another by setting the stacked parameter to True

```python
plt.title('distribution of sepal width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```
